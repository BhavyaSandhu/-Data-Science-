```python
def plot_gdp(df_economy):
    """
    Generating a bar graph 
    """
    plt.figure(figsize=(10, 6))
    plt.bar(df_economy['Year'], df_economy['GDP per capita (current US$)'], color='blue')
    plt.xlabel('Year')
    plt.ylabel('GDP per capita (current US$)')
    plt.title('GDP per capita (current US$)')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

    plt.show()
    return
    
```

```python
def plot_correlation(df_economy1, method):
    """
    Generate a heatmap to visualize the correlation between 
    
    are some of columns names from data set
    """
    fig, ax = plt.subplots(dpi=144)
   
    # Mask is used here to prevent the repetition of value
    mask = np.triu(np.ones_like(df_economy1.corr()))
    sns.heatmap(df_economy1.corr(method=method), ax=ax, vmin=-1, vmax=1, cmap='RdBu', annot=True, mask=mask)
    plt.title(method.capitalize())
    plt.show()
    return
```
